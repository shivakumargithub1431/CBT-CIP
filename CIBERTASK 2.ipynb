```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV files
file_path_1 = 'C:/Users/shiva/OneDrive/Documents/CIBERTECH 2.csv'
file_path_2 = 'C:/Users/shiva/OneDrive/Documents/CIBERTECH 3.csv'

data1 = pd.read_csv(file_path_1)
data2 = pd.read_csv(file_path_2)

# Merge the datasets
data = pd.concat([data1, data2], ignore_index=True)

# Remove duplicate rows
data = data.drop_duplicates()

# Convert 'Date' to datetime format
data['Date'] = pd.to_datetime(data['Date'], format='%d-%m-%Y')

# Define the COVID-19 period (e.g., starting from March 2020)
covid_start_date = pd.to_datetime('2020-03-01')

# Add the 'COVID Impact' column
data['COVID Impact'] = data['Date'] >= covid_start_date

# Plot Unemployment Rate over time with COVID impact highlighted
plt.figure(figsize=(12, 6))
sns.lineplot(data=data, x='Date', y='Estimated Unemployment Rate (%)', hue='COVID Impact')
plt.title('Unemployment Rate Over Time')
plt.xlabel('Date')
plt.ylabel('Unemployment Rate (%)')
plt.xticks(rotation=45)
plt.legend(title='COVID Impact', labels=['Before COVID-19', 'During COVID-19'])
plt.tight_layout()
plt.show()
```
